<h1>Water Vapor Formulas</h1>

```python
from numpy import exp, log, sqrt
```

These are some essential constants:

```python
e0 = 6.113 # hPa
T0 = 273.15 # 0°C in K
Lv = 2.5e6 # J/kg
Rv = 461 # J/kg/K
epsilon = 0.622 # ratio of gas constants for dry air and water vapor
```

Set the tmeperature, dew point, and pressure

```python
T = 15 # Celsius
Tdew = 10 # Celsius
p = 1013.25 # hPa
```

Saturation vapor pressure
$$e_s\approx e_oexp\left[\frac{L_v}{R_v}\left(\frac{1}{T_o}-\frac{1}{T}\right)\right]$$

```python
Tkel = T + T0 # Convert T from C to K
esat = e0*exp(Lv/Rv*(1/T0 - 1/Tkel))
```

Ambient vapor pressure uses the Dew Point Temperature instead of Ambient Temperature

```python
Tdkel = Tdew + T0
e = e0*exp(Lv/Rv*(1/T0 - 1/Tdkel))
```

Mixing Ratio: Ratio of vapor mass to all other components (md = mass dry air)
$$ w=\ \frac{Mass\ H2O}{Mass\ Dry\ Air}=\frac{m_v}{m_d}$$
Specic Humidity: Ratio of vapor mass to total air mass
$$q=\ \frac{Mass\ H2O}{Total\ Mass\ Air}=\frac{m_v}{m_d+m_v}=\frac{w}{1+w}$$

From the ideal gas law, we can convert $m$ to $p$ ($pV=mRT$) which gives
$$w=\epsilon\frac{e}{P-e}$$
$$q\approx\epsilon\frac{e}{P}$$
Where $\epsilon=\frac{R_d}{R_v}=0.622$


```python
w = epsilon*e/(p-e)
q = epsilon*e/p
```

For saturation Mixing Ratio and Specific Humidity, use the saturation vapor pressure, $e_{s}$

```python
wsat = epsilon*esat/(p-esat)
qsat = epsilon*esat/p
```

Relative Humidty in percent: 
$$RH=e/e_s*100 = w/w_s*100=q/q_s*100$$

```python
RH = e/esat*100
```

Print results

```python
print("Saturation vapor pressure = {:.1f} hPa".format(esat))
print("Ambient vapor pressure = {:.1f} hPa".format(e))

print("Relative humidity = {:.1f} %".format(RH))

print("Mixing ratio = {:.3g} kg/kg = {:.3g} g/kg".format(w, w*100))
print("Specific humidity = {:.3g} kg/kg = {:.3g} g/kg".format(q, q*100))

print("Saturation mixing ratio = {:.3g} kg/kg = {:.3g} g/kg".format(wsat, wsat*100))
print("Saturation specific humidity = {:.3g} kg/kg = {:.3g} g/kg".format(qsat, qsat*100))
```
